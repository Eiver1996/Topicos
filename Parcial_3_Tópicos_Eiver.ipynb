# Análisis de Citas de Control de Crecimiento y Desarrollo en Menores

## Introducción

La motivación de este trabajo es analizar y predecir si los menores de edad reciben citas de control de crecimiento y desarrollo, usando variables sociodemográficas y familiares.  
Los algoritmos implementados buscan responder:  
- ¿Cuáles menores no han recibido citas?  
- ¿Cuántas citas suelen recibir?  


Se seleccionaron las variables de entrada (P51, P52, P55, P56, P57, P6159s8, P6163s8) y se codificaron como variables dummy para que los modelos pudieran procesar datos categóricos.

Para la tarea de clasificación de la variable P6161, se dividió el dataset en conjuntos de entrenamiento y validación (80%-20%) y se entrenaron varios modelos:

    Red neuronal multicapa (MLPClassifier)

    Bosque aleatorio (RandomForestClassifier)

    Clasificador Naive Bayes (GaussianNB)

    Máquina de vectores soporte con kernel RBF (svm.SVC)

Se evaluó la precisión y matriz de confusión para comparar el desempeño de cada modelo.

Para la regresión de la variable cuantitativa P6161s1, se eliminaron las filas con valores faltantes y se entrenó un modelo MLPRegressor. Se evaluó el desempeño mediante el error cuadrático medio y se analizó la curva de pérdida para verificar el ajuste del modelo.

Finalmente, se usaron las predicciones del mejor modelo para identificar menores que no recibieron citas o que recibieron 1 o 2 citas, facilitando el análisis y la interpretación de los resultados.

```python

from google.colab import files # Para cargar archivos en google colab
uploaded = files.upload()
```

```python
import pandas as pd
import numpy as np
import io
datos_menores = pd.read_csv(io.BytesIO(uploaded['D. MENORES.txt']), delimiter='\t')
datos_menores.head(4)
```

```python

datos_menores.P52.isnull()
```

```python
datos_menores.loc[datos_menores.P52.notnull(), "P52"]
```

```python
tabla_menores = datos_menores[["P51", "P52", "P55", "P56", "P57", "P6159s8", "P6163s8", "P6161", "P6161s1"]]
tabla_menores1 = tabla_menores[tabla_menores.P52.notnull()]
tabla_menores1.head(4)
```

```python
tabla_menores1.P52.value_counts()
```

```python

tabla_menores1.P6159s8.isnull() # Las madres que no hacen ninguna actividad con los menores
tabla_menores1.loc[tabla_menores1.P6159s8.isnull(), "P6159s8"] = 2 # 2 denota las madres que sí hacen actividades con los hijos
tabla_menores1.P6163s8.isnull() # Los padres que no hacen ninguna actividad con los menores
tabla_menores1.loc[(tabla_menores.P6163s8.isnull()), "P6163s8"] = 2 # 2 denota los padres que sí hacen actividades con los hijos
tabla_menores1.P6163s8.value_counts()
```

```python

dummi_menores = pd.get_dummies(tabla_menores1, columns=["P51", "P52", "P55", "P56", "P57", "P6159s8", "P6163s8"], dtype = float)
dummi_menores.head(5)
```

```python
tabla_menores1.P6161.value_counts()
```

```python
dummi_menores.columns
```

```python
dummi_menores.rename(columns = {'P51_2':'acu casa', 'P51_3':'acu trab', 'P51_4':'niñera', 'P51_5':"mayor", 'P51_6':"menor", 'P51_7':"solo", 'P51_8':"otra persona"}, inplace = True)
dummi_menores.rename(columns = {'P52_1.0':"lejos", 'P52_2.0':"costoso", 'P52_3.0':"sin cupo", 'P52_4.0':"casa", 'P52_5.0':"edad", 'P52_6.0':"otro motivo"}, inplace = True)
dummi_menores.rename(columns = {'P55_1':"si desa", 'P55_2':"no desa", 'P56_1':"si alm", 'P56_2':"no alm", 'P57_1':"si onces", 'P57_2':"no onces"}, inplace = True)
dummi_menores.rename(columns = {'P6159s8_1.0':"no comp madre", 'P6159s8_2.0':"si comp madre", 'P6163s8_1.0':"no comp padre", 'P6163s8_2.0':"si comp padre"}, inplace = True)
dummi_menores.head(5)
```

```python
# "P51", "P52", "P55", "P56", "P57", "P6159s8" y "P6163s8", y, la salida es la variable "P6161"
x = dummi_menores.drop(["P6161", 'P6161s1'], axis = 1)
y = dummi_menores["P6161"].astype(str)
y.loc[(y == "1")] = "Si"
y.loc[(y == "2")] = "No"
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
trainX, testX, trainY, testY = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.neural_network import MLPClassifier
mlp_clf = MLPClassifier(hidden_layer_sizes=(50,50,50, 50, 50, 50, 50),
                        max_iter = 300, activation = 'logistic',
                        solver = 'sgd') # Curva de pérdida adam o sgd
```

```python
mlp_clf.fit(trainX, trainY) # Estimación de los parámetros sobre los datos de entrenamiento
y_pred = mlp_clf.predict(testX) # Predicción sobre los datos de validación (comparar la predicción con los datos de validación)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy: {:.2f}'.format(accuracy_score(y_pred, testY)))
```

```python
from sklearn.metrics import confusion_matrix
import pylab as pl
cm = confusion_matrix(testY, y_pred)
cm
import seaborn as sns
import matplotlib.pyplot as plt
ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation
# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix')
```

```python
plt.plot(mlp_clf.loss_curve_)
plt.title("Loss Curve", fontsize=14)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest.fit(trainX, trainY)
y_pred_random = random_forest.predict(testX)
print('Accuracy: {:.2f}'.format(accuracy_score(y_pred_random, testY)))
```

```python
from sklearn.metrics import confusion_matrix
import pylab as pl
cm1 = confusion_matrix(testY, y_pred_random)
cm1
import seaborn as sns
import matplotlib.pyplot as plt
ax= plt.subplot()
sns.heatmap(cm1, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation
# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix')
```

```python
from sklearn.naive_bayes import GaussianNB
clase_bayes = GaussianNB()
clase_bayes.fit(trainX, trainY)
y_pred_bayes = clase_bayes.predict(testX)
print('Accuracy: {:.2f}'.format(accuracy_score(y_pred_bayes, testY)))
```

```python
from sklearn.metrics import confusion_matrix
import pylab as pl
cm2 = confusion_matrix(testY, y_pred_bayes)
cm2
import seaborn as sns
import matplotlib.pyplot as plt
ax= plt.subplot()
sns.heatmap(cm2, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation
# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix')
```

```python
from sklearn.svm import SVC
soporte_vect = SVC(decision_function_shape='ovr', kernel = "rbf", degree = 3)
soporte_vect.fit(trainX, trainY)
y_pred_sopor = soporte_vect.predict(testX)
print('Accuracy: {:.2f}'.format(accuracy_score(y_pred_sopor, testY)))
```

```python
from sklearn.metrics import confusion_matrix
import pylab as pl
cm2 = confusion_matrix(testY, y_pred_sopor)
cm2
import seaborn as sns
import matplotlib.pyplot as plt
ax= plt.subplot()
sns.heatmap(cm2, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation
# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix')
```

Resultados

Los modelos de clasificación entrenados para predecir si un menor ha recibido citas de control (P6161) mostraron un desempeño variable. La red neuronal multicapa (MLPClassifier) logró una precisión cercana al 79%, con una matriz de confusión que refleja un balance adecuado. El modelo Random Forest también presentó una precisión similar. El clasificador Naive Bayes mostró menor precisión. La SVM con kernel RBF alcanzó un desempeño parecido.


Análisis

Los resultados sugieren que las variables sociodemográficas y familiares consideradas son relevantes para predecir la asistencia a citas de control. El tratamiento especial de valores faltantes en P6159s8 y P6163s8 mejoró la calidad de los datos, ya que estos NaN representaban información valiosa sobre la participación de los padres.

La red neuronal y el Random Forest demostraron ser los modelos más efectivos. La menor precisión de Naive Bayes indica que la independencia condicional no se cumple para estas variables.

En la regresión, la capacidad del MLPRegressor para modelar el número exacto de citas es útil para distinguir menores que no asistieron de aquellos con pocas citas, lo cual es relevante para el seguimiento y políticas de salud pública.

Estos modelos pueden apoyar a los profesionales en la identificación temprana de menores en riesgo de no recibir atención adecuada, facilitando intervenciones focalizadas.
